## Data types

This data columns in the orders table are currently stored as object values.

These columns will be converted to datetime during the data cleaning process.

Price and freight_value are stored as numerical values.

```python
for name, df in datasets.items():
    print(f"\n{name.title()}")

    missing_values = df.isnull().sum()
    missing_values = missing_values[missing_values > 0]

    if missing_values.empty:
        print("No missing values")
    else:
        print(missing_values.sort_values(ascending=False))
```

```python
missing_summary = []

for name, df in datasets.items():
    for column in df.columns:
        missing_count = df[column].isnull().sum()

        if missing_count > 0:
            missing_summary.append({
                "datasets": name,
                "column": column,
                "missing_count": missing_count,
                "missing_percentage": round(missing_count / len(df) * 100, 2)
            })

missing_summary_df = pd.DataFrame(missing_summary)

missing_summary_df.sort_values(by="missing_percentage", ascending=False)
```

```python
duplicate_summary = []

for name, df in datasets.items():
    duplicate_count = df.duplicated().sum()

    duplicate_summary.append({
        "dataset": name,
        "duplicate_row": duplicate_count
    })

duplicate_summary_df = pd.DataFrame(duplicate_summary)

duplicate_summary_df
```

## Data Quality Summary

### Dataset coverage

- The dataset covers purchases from 2016-09-04 to 2018-10-17.
- It contains 99,441 orders.
- It contains 99,441 unique customers.
- It contains 32,951 orders.

### Missing values

- Some orders have missing delivery dates.
- Some products havemissing product category names.
- Many reviews do not have written comments.
- Missing review comments do not prevent review score analysis. 

### Duplicate records

- 0 fully duplicated rows were identified.
- Repeated order IDs in the order items table are expected because one order can contain multiple products.

### Data type issues

- Order date columns are currently stored as text.
- Date columns need to be converted to datetime.
- Product category names need to be translated to English.

### Decision for data cleaning

- Keep the orginal raw files unchanged.
- Convert date fields to datetime.
- Create delivery time and delivery delay fields.
- Translate product category names.
- Use customer_unique_id for repeat purchase analysis.
- Review non-delivered orders before product and delivery analysis.
